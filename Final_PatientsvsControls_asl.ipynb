```python
#Try different percentiles of top features according to F-value and plot to identify percentile of features that provide optimal accuracy

from sklearn.svm import SVC
svc = SVC(kernel='linear', class_weight='balanced')

from sklearn.feature_selection import SelectPercentile, f_classif
feature_selection = SelectPercentile(f_classif)

from sklearn.pipeline import Pipeline
anova_svc = Pipeline([('anova', feature_selection), ('svc', svc)])
#feature selection using anova

score_means = list()
score_stds = list()
percentiles = (1, 2.5, 5, 10, 15, 20, 25, 30, 40, 50, 60, 70, 80, 90, 100) #the percentile of voxels 

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold #keeps the same proportion from each dataset.

crossval = StratifiedKFold(n_splits=5)


for percentile in percentiles:
    anova_svc.set_params(anova__percentile=percentile)
    this_scores = cross_val_score(anova_svc, asl_masked, y, cv=crossval, scoring ='balanced_accuracy')
    score_means.append(this_scores.mean())
    score_stds.append(this_scores.std())

import numpy as np
import matplotlib.pyplot as plt    
plt.errorbar(percentiles, score_means, np.array(score_stds))
plt.title(
    'Performance of the SVM-Anova varying the percentile of features selected')
plt.xticks(np.linspace(0, 100, 11, endpoint=True))
plt.xlabel('Percentile')
plt.ylabel('Accuracy Score')
plt.axis('tight')
plt.show()
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear', class_weight='balanced', probability=True) #telling the classsifier to have less weight on patient group as we have more than HC.

from sklearn.feature_selection import SelectPercentile, f_classif
feature_selection = SelectPercentile(f_classif, percentile=30)

from sklearn.pipeline import Pipeline
anova_svc = Pipeline([('anova', feature_selection), ('svc', svc)])
```

```python
#Plot ROC curves
import numpy as np
import matplotlib.pyplot as plt
from scipy import interp
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import StratifiedKFold

cv = StratifiedKFold(n_splits=5)

tprs = []
aucs = []
mean_fpr = np.linspace(0, 1, 100)

yNP=np.asarray(y)

i = 0
for train, test in cv.split(asl_masked, yNP):
    probas_= anova_svc.fit(asl_masked[train], yNP[train]).predict_proba(asl_masked[test])
    
    # Compute ROC curve and area the curve
    fpr, tpr, thresholds = roc_curve(yNP[test], probas_[:, 1])
    tprs.append(interp(mean_fpr, fpr, tpr))
    tprs[-1][0] = 0.0
    roc_auc = auc(fpr, tpr)
    aucs.append(roc_auc)
    #plt.plot(fpr, tpr, lw=1, alpha=0.3,
             #label='ROC fold %d (AUC = %0.2f)' % (i, roc_auc))

    i += 1
plt.plot([0, 1], [0, 1], linestyle='--', lw=2, color='r',
         label='Chance', alpha=.8)

mean_tpr = np.mean(tprs, axis=0)
mean_tpr[-1] = 1.0
mean_auc = auc(mean_fpr, mean_tpr)
std_auc = np.std(aucs)
plt.plot(mean_fpr, mean_tpr, color='b',
         label=r'Mean ROC (AUC = %0.2f $\pm$ %0.2f)' % (mean_auc, std_auc),
         lw=2, alpha=.8)

std_tpr = np.std(tprs, axis=0)
tprs_upper = np.minimum(mean_tpr + std_tpr, 1)
tprs_lower = np.maximum(mean_tpr - std_tpr, 0)
plt.fill_between(mean_fpr, tprs_lower, tprs_upper, color='grey', alpha=.2,
                 label=r'$\pm$ 1 std. dev.')

plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating curve')
plt.legend(loc="lower right")
plt.show()
```
